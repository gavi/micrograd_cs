Lets first define a function

$f(x) = 3x^2 + 4x + 3$

```python
def f(x):
    return 3*x**2 + 4*x + 3
```

Lets plot this

```python
xs = np.arange(-5,5,0.25)
ys = f(xs)

plt.plot(xs,ys)
plt.scatter(x=0,y=f(2),color='red')
```
